```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
traindf = pd.read_csv('Pre-processed.csv')
```

```python
traindf
```

```python
columns = traindf.columns
```

```python
normalized_df = traindf
```

```python
for col in columns:
    if col == "emission" or col == "latitude" or col == "longitude" or col == "year" or col == "week_no":
        print()
    else:
        print(col)
        normalized_df[col] = (normalized_df[col] - normalized_df[col].mean())/normalized_df[col].std()

# normalized_df
```

```python
normalized_df
```

```python
for col in columns:
    if col == "emission" or col == "latitude" or col == "longitude" or col == "year" or col == "week_no":
        print()
    else:
        print(col)
        print('{:.13f}'.format(traindf[col].mean()))
        print('{:.13f}'.format(traindf[col].std()))
```

```python
normalized_df['year'] = normalized_df['year'] .astype('Int64')
```

```python
normalized_df['week_no'] = normalized_df['week_no'] .astype('Int64')
```

```python
normalized_df.to_csv("normalised_df.csv", index= False)
```
